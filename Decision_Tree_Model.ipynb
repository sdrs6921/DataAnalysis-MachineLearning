```python
"""
Using
DecisionTreeClassifier API
learning and Visualizing Decision Tree
"""
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import export_graphviz
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
import graphviz
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
"""
Make Decision Tree ClassFier and
Learning Iris Data set Decision Tree
"""
dt_clf = DecisionTreeClassifier(random_state=156)

iris = load_iris()
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target,
                                                   test_size = 0.2, random_state= 11)
dt_clf.fit(x_train, y_train)
```

```python
export_graphviz(dt_clf, out_file="tree.dot", class_names=iris.target_names,\
               feature_names=iris.feature_names, impurity=True, filled=True)
```

```python
with open('/Users/heesang/tree.dot') as f:
    dot_graph = f.read()
    
graphviz.Source(dot_graph)
```

```python
df_clf = DecisionTreeClassifier(random_state=156, max_depth=3)

iris = load_iris()
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target,
                                                   test_size = 0.2, random_state= 11)
df_clf.fit(x_train, y_train)

export_graphviz(df_clf, out_file="tree.dot", class_names=iris.target_names,\
               feature_names=iris.feature_names, impurity=True, filled=True)
with open('/Users/heesang/tree.dot') as f:
    dot_graph = f.read()
    
graphviz.Source(dot_graph)
```

```python
df_clf = DecisionTreeClassifier(random_state=156, min_samples_leaf = 4)

iris = load_iris()
x_train, x_test, y_train, y_test = train_test_split(iris.data, iris.target,
                                                   test_size = 0.2, random_state= 11)
df_clf.fit(x_train, y_train)

export_graphviz(df_clf, out_file="tree.dot", class_names=iris.target_names,\
               feature_names=iris.feature_names, impurity=True, filled=True)
with open('/Users/heesang/tree.dot') as f:
    dot_graph = f.read()
    
graphviz.Source(dot_graph)
```

```python
print("Feature importance:\n{0}"
     .format(np.round(dt_clf.feature_importances_, 3)))

for name, value in zip(iris.feature_names, dt_clf.feature_importances_):
    print('{0} : {1:.3f}'.format(name, value))
    
sns.barplot(x=dt_clf.feature_importances_, y=iris.feature_names)
```
